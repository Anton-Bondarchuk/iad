```python
# Task 12: plot 3d graph
import numpy as np
import matplotlib.pylab as plt 


def vrosenbrock(x, y):
    return 100 * (y - x**2)**2 + (1 - x)**2

Lx, Rx, stepx = -5, 5, 0.05
Ly, Ry, stepy = -5, 5, 0.05

xs = np.arange(Lx, Rx, stepx)
ys = np.arange(Ly, Ry, stepy)
X, Y = np.meshgrid(xs, ys)

Z = vrosenbrock(X, Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Функция Розенброка")

plt.show()

```

```python
import matplotlib.pylab as plt 

X_left, X_right = -2, 2
Y_left, Y_right = -3, 3
N = 1000

def my_func(X_left, X_right, Y_left, Y_right, N):
    X = X_left + np.random.rand(N) * (X_right - X_left)
    Y = Y_left + np.random.rand(N) * (Y_right - Y_left)
    
    return X, Y

X, Y = my_func(X_left, X_right, Y_left, Y_right, N)


plt.figure()
plt.plot(X, Y, '*')
plt.title("Random points")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

BinNumber = 20
k = np.arange(BinNumber + 1)

X_bins = X_left + k * (X_right - X_left) / BinNumber
Y_bins = Y_left + k * (Y_right - Y_left) / BinNumber

N_X, _ = np.histogram(X, bins=X_bins)
N_Y, _ = np.histogram(Y, bins=Y_bins)


plt.figure()
plt.bar(X_bins[:-1], N_X, width=(X_right - X_left) / BinNumber, align='edge')
plt.title("Histogram of X")
plt.xlabel("X")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.bar(Y_bins[:-1], N_Y, width=(Y_right - Y_left) / BinNumber, align='edge')
plt.title("Histogram of Y")
plt.xlabel("Y")
plt.ylabel("Frequency")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_gauss_gen(N, m, D):
    alpha = np.random.randn(N)
    beta = m + alpha * np.sqrt(D)
    return beta

N = int(1e5)
m = 0
D = 2

data = my_gauss_gen(N, m, D)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=40, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title("Гистограмма случайных величин с гауссовским распределением")
plt.xlabel("Значение")
plt.ylabel("Частота")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Параметры
n = 3  # Размер матриц
np.random.seed(42)  # Для воспроизводимости

# Генерация случайных матриц
A = np.random.rand(n, n)
B = np.random.rand(n, n)

# Проверка свойств
def check_properties(A, B, tol=1e-6):
    results = {}
    
    # 1. (A + B)^T = A^T + B^T
    results[1] = np.allclose((A + B).T, A.T + B.T, rtol=tol, atol=tol)
    
    # 2. (AB)^T = B^T A^T
    results[2] = np.allclose((A @ B).T, B.T @ A.T, rtol=tol, atol=tol)
    
    # 3. det(A) = det(A^T)
    results[3] = np.isclose(np.linalg.det(A), np.linalg.det(A.T), rtol=tol, atol=tol)
    
    # 4. det(AB) = det(A)det(B)
    det_AB = np.linalg.det(A @ B)
    det_A_det_B = np.linalg.det(A) * np.linalg.det(B)
    results[4] = np.isclose(det_AB, det_A_det_B, rtol=tol, atol=tol)
    
    # 5. det(A^{-1}) = 1/det(A) (только для обратимой A)
    try:
        A_inv = np.linalg.inv(A)
        results[5] = np.isclose(np.linalg.det(A_inv), 1/np.linalg.det(A), rtol=tol, atol=tol)
    except np.linalg.LinAlgError:
        results[5] = "A singular"
    
    # 6. (AB)^{-1} = B^{-1}A^{-1} (только для обратимых A и B)
    try:
        AB_inv = np.linalg.inv(A @ B)
        B_inv_A_inv = np.linalg.inv(B) @ np.linalg.inv(A)
        results[6] = np.allclose(AB_inv, B_inv_A_inv, rtol=tol, atol=tol)
    except np.linalg.LinAlgError:
        results[6] = "Matrix singular"
    
    # 7. (A - B)^{-1} = A^{-1} + A^{-1}B(A - B)^{-1} (условно)
    try:
        A_minus_B = A - B
        lhs = np.linalg.inv(A_minus_B)
        A_inv = np.linalg.inv(A)
        rhs = A_inv + A_inv @ B @ np.linalg.inv(A_minus_B)
        results[7] = np.allclose(lhs, rhs, rtol=tol, atol=tol)
    except np.linalg.LinAlgError:
        results[7] = "Matrix singular"
    
    return results

# Выполнение проверки
properties = check_properties(A, B)

# Вывод результатов
print(f"Матрица A:\n{A}\n")
print(f"Матрица B:\n{B}\n")
for prop, result in properties.items():
    print(f"Свойство {prop}: {'Выполнено' if result is True else 'Не выполнено' if result is False else result}")
```
